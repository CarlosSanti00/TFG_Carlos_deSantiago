# Comparison of the estimated parameters

```python
import matplotlib.pyplot as plt 
import matplotlib.dates as mdates
%matplotlib inline
import pandas as pd
import numpy as np
from IPython import display
import datetime as dt
```

```python
# Loading parameters estimated from ODEs
df_ODE = pd.read_csv('Parameters_ODEs.csv')
df_ODE
```

```python
# Obtaining dates and parameters
dates_ODE = df_ODE['Date'].values
beta_ODE = df_ODE['Beta Fixed Smooth'].values
gamma_ODE = df_ODE['Gamma Smooth (7 days average)'].values
mu_ODE = df_ODE['Mu Smooth (7 days average)'].values

len(dates_ODE), len(beta_ODE), len(gamma_ODE), len(mu_ODE)
```

```python
# Loading parameters estimated from LM
df_LM = pd.read_csv('Parameters_LM.csv')
df_LM
```

```python
# Obtaining dates and parameters
dates_LM = df_LM['Date'].values
beta_LM = df_LM['Beta Fixed Smooth'].values
gamma_LM = df_LM['Gamma Fixed Smooth'].values
mu_LM = df_LM['Mu Fixed Smooth'].values

len(dates_LM), len(beta_LM), len(gamma_LM), len(mu_LM)
```

```python
# Adjustin gamma LM value when 0 values

for i in range(0, len(gamma_LM), 1):
    if (gamma_LM[i] < 0):
        gamma_LM[i] = 0 
```

```python
# Checking if both dataframes begin in the same day
print(str(dates_ODE[0] == dates_LM[0]))
```

```python
# Adjusting the length of parameters to have the same length
dates_ODE = dates_ODE[0:len(dates_LM)]
beta_ODE = beta_ODE[0:len(dates_LM)]
gamma_ODE = gamma_ODE[0:len(dates_LM)]
mu_ODE = mu_ODE[0:len(dates_LM)]

len(dates_ODE), len(beta_ODE), len(gamma_ODE), len(mu_ODE)
```

```python
# Check if the 'Date' data from both datasets are equal
print(all(dates_ODE == dates_LM))
```

```python
# Transform 'Date' string format to date format
date_label = []
for i in range(0, len(dates_LM), 1):
    date = dt.datetime.strptime(dates_LM[i],'%Y-%m-%d').date()
    date_label.append(date.strftime('%d/%m/%y')) 

len(date_label)
```

```python
# Function to plot beta
def plot_beta(beta, date_label, color, x_cord, title, download):
    
    # Importing the style package
    from matplotlib import style
    # Using the style for the plot
    plt.style.use('ggplot')
    
    plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=3))
    plt.plot(date_label, beta, linewidth = 1, color = color)
    plt.ylabel('Beta [1/dÃ­a]', size = 'medium')
    plt.yticks(fontsize = 'small')
    plt.xlabel('Tiempo (dÃ­as)', size = 'medium')
    plt.xticks(fontsize = 'small')
    plt.suptitle(title, y = 0, x = x_cord, size = 'large')
    plt.gcf().autofmt_xdate()
    plt.xlim(left=-10, right=len(date_label)+10)
    
    # Saving plot
    name_file = download + '.png'
    plt.tight_layout()
    plt.savefig(name_file, format='png', dpi=1200, facecolor='white', bbox_inches='tight')
```

```python
# Beta ODEs
plot_beta(beta_ODE, date_label, color = 'cornflowerblue', 
          x_cord = 0.54, title = '(a) EspaÃ±a - Î²(t) ODEs', download = 'Beta_ODEs')
```

```python
# Beta LM
plot_beta(beta_LM, date_label, color = 'slateblue', x_cord = 0.54, title = '(b) EspaÃ±a - Î²(t) LM', download = 'Beta_LM')
```

```python
# Function to plot beta
def plot_gamma(gamma, date_label, color, x_cord, title, download):
    
    # Importing the style package
    from matplotlib import style
    # Using the style for the plot
    plt.style.use('ggplot')
    
    plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=3))
    plt.plot(date_label, gamma, linewidth = 1, label = 'Î³(t)', color = color)
    plt.axhline(y=1/14, color='black', linestyle='--', linewidth = 0.75, label = "Î³ = 1/14")
    plt.axhline(y=np.mean(gamma), color='deeppink', linestyle='--', linewidth = 0.75, label = "Î³(t) promedio")
    plt.ylabel('Gamma [1/dÃ­a]', size = 'medium')
    plt.yticks(fontsize = 'small')
    plt.xlabel('Tiempo (dÃ­as)', size = 'medium')
    plt.xticks(fontsize = 'small')
    plt.suptitle(title, y = 0, x = x_cord, size = 'large')
    plt.legend(loc=(0.65, 0.73), fontsize = 'medium')
    plt.gcf().autofmt_xdate()
    plt.xlim(left=-10, right=len(date_label)+10)
    
    # Saving plot
    name_file = download + '.png'
    plt.tight_layout()
    plt.savefig(name_file, format='png', dpi=1200, facecolor='white', bbox_inches='tight')
```

```python
# Gamma ODEs
plot_gamma(gamma_ODE, date_label, color = 'olivedrab', 
           x_cord = 0.545, title = '(a) EspaÃ±a - Î³(t) ODEs', download = 'Gamma_ODEs')
```

```python
# Gamma LM
plot_gamma(gamma_LM, date_label, color = 'mediumseagreen',
           x_cord = 0.545, title = '(b) EspaÃ±a - Î³(t) LM', download = 'Gamma_LM')
```

```python
# Function to plot beta
def plot_mu(mu, date_label, color, x_cord, title, download):
    
    # Importing the style package
    from matplotlib import style
    # Using the style for the plot
    plt.style.use('ggplot')
    
    plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=3))
    plt.plot(date_label, mu, linewidth = 1, color = color)
    plt.ylabel('Mu [1/dÃ­a]', size = 'medium')
    plt.yticks(fontsize = 'small')
    plt.xlabel('Tiempo (dÃ­as)', size = 'medium')
    plt.xticks(fontsize = 'small')
    plt.suptitle(title, y = 0, x = x_cord, size = 'large')
    plt.gcf().autofmt_xdate()
    plt.xlim(left=-10, right=len(date_label)+10)
    
    # Saving plot
    name_file = download + '.png'
    plt.tight_layout()
    plt.savefig(name_file, format='png', dpi=1200, facecolor='white', bbox_inches='tight')
```

```python
# Mu ODEs
plot_mu(mu_ODE, date_label, color = 'indianred', 
        x_cord = 0.55, title = '(a) EspaÃ±a - Î¼(t) ODEs', download = 'Mu_ODEs')
```

```python
# Mu LM
plot_mu(mu_LM, date_label, color = 'tomato',
        x_cord = 0.545, title = '(b) EspaÃ±a - Î¼(t) LM', download = 'Mu_LM')
```
